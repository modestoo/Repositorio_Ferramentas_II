```python
# Import das principais bibliotecas
import pandas as pd
```

```python
# Criando o dataframe a partir do xlsx
df = pd.read_excel("201710_EmissaoPassagens_SCDP.xlsx", skiprows=3, skipfooter=3)
```

```python
# Removendo os espaços adicionais que vieram no excel
df.rename(columns={' Código do órgão superior': 'Código do órgão superior',
                   '                               Nome do órgão superior                                ':'Nome do órgão superior',
                   ' Código do órgão solicitante da viagem':'Código do órgão solicitante da viagem',
                   '                                      Nome do órgão solicitante da viagem                                       ':'Nome do órgão solicitante da viagem',
                   '    N. PCDP    ':'N. PCDP',
                   ' N. Reserva/Localizador':'N. Reserva/Localizador',
                   '  Data Emissão Bilhete   ':'Data Emissão Bilhete',
                   '    Data Embarque    ':'Data Embarque',
                   ' Valor Tarifa Comercial':'Valor Tarifa Comercial',
                   ' Percentual Desconto Aplicado':'Percentual Desconto Aplicado',
                   ' Valor Tarifa Governo':'Valor Tarifa Governo',
                   ' Valor Tarifa Embarque':'Valor Tarifa Embarque',
                   ' Valor Bilhete':'Valor Bilhete',
                   '           Companhia Aérea            ':'Companhia Aérea',
                   ' Classe Tarifária Bilhete':'Classe Tarifária Bilhete',
                   '                                       Regra Tarifária                                       ':'Regra Tarifária',
                   ' No Show':'No Show',
                   ' Remarcado':'Remarcado',
                   ' Cancelado':'Cancelado',
                   ' Valor Multas':'Valor Multas',
                   ' Valor Reembolso':'Valor Reembolso',
                   ' Diferença de Tarifa':'Diferença de Tarifa',
                   ' Situação Final Bilhete':'Situação Final Bilhete'                     
                  }, inplace=True)
```

```python
# Verificando se os nomes das colunas foram alterados
df.info()
```

```python
# Removendo os espaços das colunas 'object'
df['Nome do órgão superior'] = df['Nome do órgão superior'].str.strip()
df['Nome do órgão solicitante da viagem'] = df['Nome do órgão solicitante da viagem'].str.strip()
df['N. PCDP'] = df['N. PCDP'].str.strip()
df['N. Reserva/Localizador'] = df['N. Reserva/Localizador'].str.strip()
df['Data Emissão Bilhete'] = df['Data Emissão Bilhete'].str.strip()
df['Data Embarque'] = df['Data Embarque'].str.strip()
df['Companhia Aérea'] = df['Companhia Aérea'].str.strip()
df['Classe Tarifária Bilhete'] = df['Classe Tarifária Bilhete'].str.strip()
df['Regra Tarifária'] = df['Regra Tarifária'].str.strip()
df['No Show'] = df['No Show'].str.strip()
df['Remarcado'] = df['Remarcado'].str.strip()
df['Cancelado'] = df['Cancelado'].str.strip()
df['Situação Final Bilhete'] = df['Situação Final Bilhete'].str.strip()
```

```python
# Verificando se há sujeira nos campos de data
df['Data Embarque'].value_counts()
```

```python
# Verificando se há sujeira nos campos de data
df['Data Emissão Bilhete'].value_counts()
```

```python
# Transformando os campos de data em tipo da datetime
df['Data Emissão Bilhete'] = pd.to_datetime(df['Data Emissão Bilhete'])
df['Data Embarque'] = pd.to_datetime(df['Data Embarque'])
```

```python
# Verificando se as datas foram alteradas
df.info()
```

```python
# Verificando a quantidade de solicitações por órgão solicitantes 
df['Nome do órgão superior'].value_counts()

# Ministério da Fazenda possui o maior número de solicitações
# Instituto de Pesquisas Jardim Botânico do Rio de Janeiro possui o menor número
```

```python
# Verificando quando o Governo deixou de gastar a partir dos 'descontos' das empresas com base na diferença
# entre as colunas valor tarifa comercial e governo
df['Desconto na Passagem'] = df['Valor Tarifa Comercial'] - df['Valor Tarifa Governo']
df['Desconto na Passagem'].sum()
```

```python
# Quais os 10 primeiros órgãos obtiveram os maiores valores de descontos
df.nlargest(10, 'Desconto na Passagem')
```

```python
# Verificando quando o governo gastou em bilhetes
df['Valor Bilhete'].sum()
```

```python
# Rankiando os 10 orgãos com maior valor em bilhetes únicos
df.nlargest(10, 'Valor Bilhete')
```

```python
# Qual a quantidade de bilhetes por companhias
%matplotlib inline
df['Companhia Aérea'].value_counts().plot.bar()
```

```python
# Quanto de multas foram pagas pelo Governo?
df['Valor Multas'].sum()
```

```python
# Os voos que foram Remarcados tiveram que pagar multa?
df_remarcada = df[df['Remarcado'] == 'Sim']
df_remarcada[['Nome do órgão superior', 'Remarcado', 'Valor Multas']]
```

```python
# Quais voos foram cancelados e não foram reembolsados?
df_cancelados = df[df['Cancelado'] == 'Sim']
df_cancelados[df_cancelados['Valor Reembolso'] == 0]
```

```python
# Qual a situação geral dos voos?
df['Situação Final Bilhete'].value_counts().plot.barh()
```

```python
# Quais os dias do mês de Outubro com maior quantidade de Emissão de Bilhetes?
df_outubro = df[df['Data Emissão Bilhete'].dt.month == 10]
df_outubro['Data Emissão Bilhete'].dt.day.value_counts().plot.bar()
```

```python
# Quais os custos totais com bilhete por Companhias?
df.groupby('Companhia Aérea')['Valor Bilhete'].agg(['sum','count'])
```
